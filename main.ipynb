```python
import pandas as pd
import numpy as np

# Load data
df = pd.read_csv('S2324-laliga-players.csv')

# Separate by competition type
df_classics = df[df['competition'] == 'Classics'].copy()
df_efficiency = df[df['competition'] == 'Efficiency'].copy()
df_discipline = df[df['competition'] == 'Discipline'].copy()
df_offensives = df[df['competition'] == 'Offensives'].copy()
df_defensives = df[df['competition'] == 'Deffensives'].copy()

print(f"Classics columns: {len(df_classics.columns)}")
print(f"Efficiency columns: {len(df_efficiency.columns)}")
print(f"Discipline columns: {len(df_discipline.columns)}")
print(f"Offensives columns: {len(df_offensives.columns)}")
print(f"Defensives columns: {len(df_defensives.columns)}")

# Identify base/metadata columns (present in all)
base_cols = ['id', 'name', 'firstname', 'lastname', 'nickname', 'position', 
             'team', 'team.shortname', 'shirt_number', 'date_of_birth', 
             'place_of_birth', 'country', 'height', 'weight', 'photo',
             'player.url', 'slug', 'gender', 'international']

# Identify unique stat columns per competition type
classics_stats = [col for col in df_classics.columns if col not in base_cols and col != 'competition']
efficiency_stats = [col for col in df_efficiency.columns if col not in base_cols and col != 'competition']
discipline_stats = [col for col in df_discipline.columns if col not in base_cols and col != 'competition']
offensives_stats = [col for col in df_offensives.columns if col not in base_cols and col != 'competition']
defensives_stats = [col for col in df_defensives.columns if col not in base_cols and col != 'competition']

print("\n=== Stat columns by competition ===")
print(f"Classics unique stats: {len(classics_stats)}")
print(f"Efficiency unique stats: {len(efficiency_stats)}")
print(f"Discipline unique stats: {len(discipline_stats)}")
print(f"Offensives unique stats: {len(offensives_stats)}")
print(f"Defensives unique stats: {len(defensives_stats)}")

# Let's see what stats are in each
print("\nFirst 20 Offensives stats:", offensives_stats[:20])
print("\nFirst 20 Defensives stats:", defensives_stats[:20])
```

```python
# Let's check which stats are actually populated (non-NaN) in each competition
def analyze_nan_distribution(df, comp_name):
    """Analyze which columns have actual data vs NaN"""
    # Exclude base metadata columns
    stat_cols = [col for col in df.columns if col not in base_cols and col != 'competition']
    
    # Calculate NaN percentage for each column
    nan_percentages = {}
    for col in stat_cols:
        nan_pct = (df[col].isnull().sum() / len(df)) * 100
        if nan_pct < 100:  # Only include columns with some data
            nan_percentages[col] = nan_pct
    
    # Sort by NaN percentage
    sorted_cols = sorted(nan_percentages.items(), key=lambda x: x[1])
    
    print(f"\n{'='*60}")
    print(f"{comp_name} - Columns with data (sorted by completeness)")
    print(f"{'='*60}")
    print(f"Total columns: {len(stat_cols)}, Columns with data: {len(sorted_cols)}")
    print("\nMost complete columns (< 50% NaN):")
    for col, pct in sorted_cols[:30]:  # Show top 30
        if pct < 50:
            print(f"  {col:<45} {pct:>6.1f}% NaN")
    
    return sorted_cols

# Analyze each competition type
classics_data = analyze_nan_distribution(df_classics, "CLASSICS")
efficiency_data = analyze_nan_distribution(df_efficiency, "EFFICIENCY")
discipline_data = analyze_nan_distribution(df_discipline, "DISCIPLINE")
offensives_data = analyze_nan_distribution(df_offensives, "OFFENSIVES")
defensives_data = analyze_nan_distribution(df_defensives, "DEFENSIVES")

# Now let's see which columns are unique to each competition (have data there but not elsewhere)
print("\n" + "="*60)
print("COMPARISON: Which stats are populated in which competition?")
print("="*60)

# Get columns with <50% NaN in each
classics_cols = {col for col, pct in classics_data if pct < 50}
efficiency_cols = {col for col, pct in efficiency_data if pct < 50}
discipline_cols = {col for col, pct in discipline_data if pct < 50}
offensives_cols = {col for col, pct in offensives_data if pct < 50}
defensives_cols = {col for col, pct in defensives_data if pct < 50}

print(f"\nClassics has {len(classics_cols)} usable columns")
print(f"Efficiency has {len(efficiency_cols)} usable columns")
print(f"Discipline has {len(discipline_cols)} usable columns")
print(f"Offensives has {len(offensives_cols)} usable columns")
print(f"Defensives has {len(defensives_cols)} usable columns")

# Check overlap
print("\nColumns ONLY in Offensives:", offensives_cols - classics_cols - efficiency_cols - discipline_cols - defensives_cols)
print("\nColumns ONLY in Defensives:", defensives_cols - classics_cols - efficiency_cols - discipline_cols - offensives_cols)
print("\nColumns ONLY in Efficiency:", efficiency_cols - classics_cols - offensives_cols - discipline_cols - defensives_cols)
```

```python
# Let's check if the same player has different values across competitions
# Pick a specific player and see their data across all competitions

sample_player_id = df_classics['id'].iloc[245]  # Get first player's ID
print(f"Analyzing player ID: {sample_player_id}")
print(f"Player name: {df[df['id'] == sample_player_id]['name'].iloc[0]}")

# Get this player's data across all competitions
player_data = df[df['id'] == sample_player_id]

print(f"\nThis player has {len(player_data)} entries (one per competition)")
print("\nCompetition types for this player:")
print(player_data['competition'].values)

# Compare key stats across competitions for this player
comparison_cols = ['competition', 'appearances', 'goals', 'goal_assists', 'total_passes', 
                   'total_shots', 'total_tackles', 'yellow_cards', 'blocks', 'saves_made']

print("\nKey stats across competitions for this player:")
print(player_data[comparison_cols].T)

print("\n" + "="*60)
print("Checking if stats are identical or different across competitions...")
print("="*60)

# Check if all 5 rows are identical for this player
for col in ['goals', 'goal_assists', 'total_passes', 'total_shots', 'saves_made', 'total_tackles', 'yellow_cards']:
    values = player_data[col].values
    if pd.notna(values).any():  # If there's any non-NaN value
        unique_values = player_data[col].dropna().unique()
        print(f"{col:<30} Unique non-NaN values: {unique_values}")

# Let's check a few more players to see the pattern
print("\n" + "="*60)
print("Checking 5 random players to confirm pattern...")
print("="*60)

for player_id in df_classics['id'].sample(5).values:
    player_name = df[df['id'] == player_id]['name'].iloc[0]
    player_pos = df[df['id'] == player_id]['position'].iloc[0]
    player_comps = df[df['id'] == player_id]
    
    # Check if goals differ across competitions
    goals_values = player_comps['goals'].dropna().unique()
    assists_values = player_comps['goal_assists'].dropna().unique()
    
    print(f"\n{player_name} ({player_pos}):")
    print(f"  Goals unique values: {goals_values}")
    print(f"  Assists unique values: {assists_values}")
```

```python
# Since all 5 competition rows are duplicates, just keep one per player
# Let's use Classics as our source

print("Original dataset shape:", df.shape)
print("Players in Classics:", len(df_classics))
print("Unique player IDs:", df_classics['id'].nunique())

# Use only Classics (one entry per player)
df_clean = df_classics.copy()

# Filter to players with at least 5 appearances
df_clean = df_clean[df_clean['appearances'].fillna(0) >= 5].copy()

print("\nAfter filtering (appearances >= 5):")
print("Shape:", df_clean.shape)
print("Unique players:", df_clean['id'].nunique())

# Fill NaN values intelligently based on position
print("\n" + "="*60)
print("Handling NaN values by position...")
print("="*60)

# Attacking stats - fill with 0 (especially for GKs/Defenders)
attacking_stats = [
    'goals', 'goal_assists', 'total_shots', 'shots_on_target_inc_goals',
    'shots_off_target_inc_woodwork', 'successful_dribbles', 'unsuccessful_dribbles',
    'headed_goals', 'left_foot_goals', 'right_foot_goals', 'penalty_goals',
    'winning_goal', 'attempts_from_set_pieces', 'hit_woodwork'
]

# Goalkeeper-specific stats - fill with 0 for non-GKs
gk_stats = [
    'saves_made', 'saves_made_from_inside_box', 'saves_made_from_outside_box',
    'catches', 'punches', 'drops', 'crosses_not_claimed', 'gk_successful_distribution',
    'gk_unsuccessful_distribution', 'goalkeeper_smother', 'goal_kicks',
    'penalties_saved', 'penalties_faced', 'clean_sheets', 'goals_conceded',
    'goals_conceded_inside_box', 'goals_conceded_outside_box', 'saves_made_caught',
    'saves_made_parried', 'saves_from_penalty', 'penalty_goals_conceded'
]

# Defensive stats - fill with 0
defensive_stats = [
    'total_tackles', 'tackles_won', 'tackles_lost', 'blocks', 'interceptions',
    'clearances_off_the_line', 'total_clearances', 'last_player_tackle'
]

# Discipline stats - fill with 0
discipline_stats = [
    'yellow_cards', 'total_red_cards', 'red_cards_2nd_yellow', 'straight_red_cards',
    'total_fouls_conceded', 'total_fouls_won', 'penalties_conceded',
    'handballs_conceded', 'own_goal_scored'
]

# Fill attacking stats with 0
for col in attacking_stats:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].fillna(0)

# Fill GK stats with 0
for col in gk_stats:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].fillna(0)

# Fill defensive stats with 0
for col in defensive_stats:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].fillna(0)

# Fill discipline stats with 0
for col in discipline_stats:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].fillna(0)

print("\nNaN values after filling:")
print(df_clean[['goals', 'goal_assists', 'total_shots', 'saves_made', 
               'total_tackles', 'yellow_cards']].isnull().sum())

# Calculate per-90 metrics for fair comparison
print("\n" + "="*60)
print("Calculating per-90 metrics...")
print("="*60)

df_clean['minutes_played'] = df_clean['time_played']
df_clean['goals_per90'] = (df_clean['goals'] / df_clean['minutes_played']) * 90
df_clean['assists_per90'] = (df_clean['goal_assists'] / df_clean['minutes_played']) * 90
df_clean['shots_per90'] = (df_clean['total_shots'] / df_clean['minutes_played']) * 90
df_clean['passes_per90'] = (df_clean['total_passes'] / df_clean['minutes_played']) * 90
df_clean['tackles_per90'] = (df_clean['total_tackles'] / df_clean['minutes_played']) * 90
df_clean['duels_won_per90'] = (df_clean['duels_won'] / df_clean['minutes_played']) * 90

# Calculate pass accuracy
df_clean['pass_accuracy'] = (df_clean['total_successful_passes_excl_crosses_corners'] / 
                             df_clean['total_passes']) * 100

# Calculate shot accuracy
df_clean['shot_accuracy'] = (df_clean['shots_on_target_inc_goals'] / 
                             df_clean['total_shots']) * 100

# Calculate duel win rate
df_clean['duel_win_rate'] = (df_clean['duels_won'] / df_clean['duels']) * 100

# Replace inf values with NaN (from division by zero)
df_clean = df_clean.replace([np.inf, -np.inf], np.nan)

print("Sample of calculated metrics:")
print(df_clean[['name', 'position', 'goals', 'goals_per90', 'pass_accuracy', 
               'duel_win_rate']].head(10))

# Save clean dataset
print("\n" + "="*60)
print("Saving cleaned dataset...")
print("="*60)

df_clean.to_csv('laliga_players_cleaned.csv', index=False)
print(f"✓ Saved: laliga_players_cleaned.csv")
print(f"  Shape: {df_clean.shape}")
print(f"  Players: {len(df_clean)}")
print(f"\nPosition distribution:")
print(df_clean['position'].value_counts())
```
